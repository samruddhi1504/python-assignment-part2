# Restaurant Order Management System - Part 2

```python
menu = {
    "Paneer Tikka":   {"category": "Starters",  "price": 180.0, "available": True},
    "Chicken Wings":  {"category": "Starters",  "price": 220.0, "available": False},
    "Veg Soup":       {"category": "Starters",  "price": 120.0, "available": True},
    "Butter Chicken": {"category": "Mains",     "price": 320.0, "available": True},
    "Dal Tadka":      {"category": "Mains",     "price": 180.0, "available": True},
    "Veg Biryani":    {"category": "Mains",     "price": 250.0, "available": True},
    "Garlic Naan":    {"category": "Mains",     "price":  40.0, "available": True},
    "Gulab Jamun":    {"category": "Desserts",  "price":  90.0, "available": True},
    "Rasgulla":       {"category": "Desserts",  "price":  80.0, "available": True},
    "Ice Cream":      {"category": "Desserts",  "price": 110.0, "available": False},
}

inventory = {
    "Paneer Tikka":   {"stock": 10, "reorder_level": 3},
    "Chicken Wings":  {"stock":  8, "reorder_level": 2},
    "Veg Soup":       {"stock": 15, "reorder_level": 5},
    "Butter Chicken": {"stock": 12, "reorder_level": 4},
    "Dal Tadka":      {"stock": 20, "reorder_level": 5},
    "Veg Biryani":    {"stock":  6, "reorder_level": 3},
    "Garlic Naan":    {"stock": 30, "reorder_level": 10},
    "Gulab Jamun":    {"stock":  5, "reorder_level": 2},
    "Rasgulla":       {"stock":  4, "reorder_level": 3},
    "Ice Cream":      {"stock":  7, "reorder_level": 4},
}
```

```python
# Create empty order list
order = []

def add_item(item_name):
    if item_name in menu:
        if menu[item_name]["available"] and inventory[item_name]["stock"] > 0:
            order.append(item_name)
            inventory[item_name]["stock"] -= 1
            print(f"{item_name} added to order")
        else:
            print(f"{item_name} is not available")
    else:
        print("Item not found")

def remove_item(item_name):
    if item_name in order:
        order.remove(item_name)
        inventory[item_name]["stock"] += 1
        print(f"{item_name} removed from order")
    else:
        print("Item not in order")

def calculate_total():
    total = 0
    for item in order:
        total += menu[item]["price"]
    return total

def show_order():
    print("\nYour Order:")
    for item in order:
        print("-", item)
    print("Total:", calculate_total())
```

```python
add_item("Paneer Tikka")
add_item("Garlic Naan")
add_item("Ice Cream")  # unavailable

show_order()

remove_item("Garlic Naan")

show_order()
```

Task 2 — Cart Operations 

```python
# Cart starts empty
cart = []

# Function to add item
def add_to_cart(item_name, quantity):
    # Check if item exists
    if item_name not in menu:
        print(f"{item_name} does not exist in menu")
        return
    
    # Check availability
    if not menu[item_name]["available"]:
        print(f"{item_name} is currently unavailable")
        return
    
    # Check if already in cart
    for item in cart:
        if item["item"] == item_name:
            item["quantity"] += quantity
            print(f"Updated {item_name} quantity to {item['quantity']}")
            return
    
    # Add new item
    cart.append({
        "item": item_name,
        "quantity": quantity,
        "price": menu[item_name]["price"]
    })
    print(f"Added {item_name} x{quantity} to cart")


# Function to remove item
def remove_from_cart(item_name):
    for item in cart:
        if item["item"] == item_name:
            cart.remove(item)
            print(f"{item_name} removed from cart")
            return
    print(f"{item_name} not found in cart")


# Function to update quantity
def update_quantity(item_name, quantity):
    for item in cart:
        if item["item"] == item_name:
            item["quantity"] = quantity
            print(f"{item_name} quantity updated to {quantity}")
            return
    print(f"{item_name} not found in cart")


# Function to print cart
def print_cart():
    print("\nCurrent Cart:")
    if not cart:
        print("Cart is empty")
    else:
        for item in cart:
            print(item)
```

```python
# Step 1
add_to_cart("Paneer Tikka", 2)
print_cart()

# Step 2
add_to_cart("Gulab Jamun", 1)
print_cart()

# Step 3
add_to_cart("Paneer Tikka", 1)
print_cart()

# Step 4
add_to_cart("Mystery Burger", 1)
print_cart()

# Step 5
add_to_cart("Chicken Wings", 1)
print_cart()

# Step 6
remove_from_cart("Gulab Jamun")
print_cart()
```

```python
def print_order_summary():
    print("\n========== Order Summary ==========")
    
    subtotal = 0
    
    for item in cart:
        item_total = item["quantity"] * item["price"]
        subtotal += item_total
        
        print(f"{item['item']:15} x{item['quantity']}    ₹{item_total:.2f}")
    
    print("------------------------------------")
    print(f"Subtotal:                ₹{subtotal:.2f}")
    
    gst = subtotal * 0.05
    print(f"GST (5%):                ₹{gst:.2f}")
    
    total = subtotal + gst
    print(f"Total Payable:           ₹{total:.2f}")
    print("====================================")

print_order_summary()
```

Task 3 — Inventory Tracker with Deep Copy

```python
import copy

# Create deep copy
inventory_backup = copy.deepcopy(inventory)

print("Original Inventory (Before Change):")
print(inventory)

print("\nBackup Inventory:")
print(inventory_backup)

# Modify original inventory (for testing deep copy)
inventory["Paneer Tikka"]["stock"] = 5

print("\nAfter modifying original inventory:")

print("Modified Inventory:")
print(inventory)

print("\nBackup Inventory (should be unchanged):")
print(inventory_backup)

# Restore original inventory
inventory = copy.deepcopy(inventory_backup)

print("\nInventory restored to original state")
```

```python
# Deduct stock based on cart

for cart_item in cart:
    item_name = cart_item["item"]
    quantity_needed = cart_item["quantity"]
    
    available_stock = inventory[item_name]["stock"]
    
    if available_stock >= quantity_needed:
        inventory[item_name]["stock"] -= quantity_needed
        print(f"{item_name}: {quantity_needed} deducted, remaining {inventory[item_name]['stock']}")
    
    else:
        print(f"⚠ Warning: Only {available_stock} {item_name} available, cannot fulfill {quantity_needed}")
        inventory[item_name]["stock"] = 0
```

```python
print("\nReorder Alerts:")

for item, details in inventory.items():
    if details["stock"] <= details["reorder_level"]:
        print(f"⚠ Reorder Alert: {item} — Only {details['stock']} unit(s) left (reorder level: {details['reorder_level']})")
```

```python
print("\nFinal Inventory:")
print(inventory)

print("\nBackup Inventory (Unchanged):")
print(inventory_backup)
```

## Task 4 — Daily Sales Log Analysis 

```python
# -------------------------------
# SALES LOG DATA (must be defined first)
# -------------------------------
sales_log = {
    "2025-01-01": [
        {"order_id": 1,  "items": ["Paneer Tikka", "Garlic Naan"],          "total": 220.0},
        {"order_id": 2,  "items": ["Gulab Jamun", "Veg Soup"],              "total": 210.0},
        {"order_id": 3,  "items": ["Butter Chicken", "Garlic Naan"],        "total": 360.0},
    ],
    "2025-01-02": [
        {"order_id": 4,  "items": ["Dal Tadka", "Garlic Naan"],             "total": 220.0},
        {"order_id": 5,  "items": ["Veg Biryani", "Gulab Jamun"],           "total": 340.0},
    ],
    "2025-01-03": [
        {"order_id": 6,  "items": ["Paneer Tikka", "Rasgulla"],             "total": 260.0},
        {"order_id": 7,  "items": ["Butter Chicken", "Veg Biryani"],        "total": 570.0},
        {"order_id": 8,  "items": ["Garlic Naan", "Gulab Jamun"],           "total": 130.0},
    ],
    "2025-01-04": [
        {"order_id": 9,  "items": ["Dal Tadka", "Garlic Naan", "Rasgulla"], "total": 300.0},
        {"order_id": 10, "items": ["Paneer Tikka", "Gulab Jamun"],          "total": 270.0},
    ],
}

# -------------------------------
# 1. Revenue Per Day
# -------------------------------
daily_revenue = {}

for date, orders in sales_log.items():
    total = 0
    for order in orders:
        total += order["total"]
    daily_revenue[date] = total

print("Revenue Per Day:")
for date, revenue in daily_revenue.items():
    print(f"{date} : ₹{revenue}")

# -------------------------------
# 2. Best Selling Day
# -------------------------------
best_day = max(daily_revenue, key=daily_revenue.get)
print(f"\nBest Selling Day: {best_day} with ₹{daily_revenue[best_day]}")

# -------------------------------
# 3. Most Ordered Item
# -------------------------------
item_count = {}

for orders in sales_log.values():
    for order in orders:
        for item in order["items"]:
            item_count[item] = item_count.get(item, 0) + 1

most_ordered = max(item_count, key=item_count.get)
print(f"\nMost Ordered Item: {most_ordered} ({item_count[most_ordered]} times)")

# -------------------------------
# 4. Add New Day
# -------------------------------
sales_log["2025-01-05"] = [
    {"order_id": 11, "items": ["Butter Chicken", "Gulab Jamun", "Garlic Naan"], "total": 490.0},
    {"order_id": 12, "items": ["Paneer Tikka", "Rasgulla"],                     "total": 260.0},
]

# -------------------------------
# 5. Updated Revenue & Best Day
# -------------------------------
daily_revenue = {}

for date, orders in sales_log.items():
    total = 0
    for order in orders:
        total += order["total"]
    daily_revenue[date] = total

print("\nUpdated Revenue Per Day:")
for date, revenue in daily_revenue.items():
    print(f"{date} : ₹{revenue}")

best_day = max(daily_revenue, key=daily_revenue.get)
print(f"\nUpdated Best Selling Day: {best_day} with ₹{daily_revenue[best_day]}")

# -------------------------------
# 6. Enumerate All Orders
# -------------------------------
print("\nAll Orders:\n")

counter = 1

for date, orders in sales_log.items():
    for order in orders:
        items = ", ".join(order["items"])
        
        print(f"{counter}. [{date}] Order #{order['order_id']} — ₹{order['total']} — Items: {items}")
        
        counter += 1
```
